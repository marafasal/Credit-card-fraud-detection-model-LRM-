```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset
fraud_data=pd.read_csv('/content/creditcard.csv')
```

```python
print(fraud_data.head())
```

```python
fraud_data.info()
```

```python
fraud_data.isnull().sum()
```

```python
#chcking the distribution of legit transaction and fraudulent transaction
fraud_data['Class'].value_counts()
```

Highly imbalanced data
0---Normal transaction
1---Fraudulent transaction

```python
legit=fraud_data[fraud_data.Class==0]
fraud=fraud_data[fraud_data.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measurement of the data
legit.Amount.describe
fraud.Amount.describe
```

```python
fraud_data.groupby("Class").mean()
```

Building an undersample datset of normal transaction and fraudulent transaction
Taking equivalence of fraudulent transaction randomly

```python
legit_sample=legit.sample(153)
```

```python
#concatenating the data
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby("Class").mean()
```

```python
#split data into features and target
X=new_dataset.drop(columns="Class",axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model training


```python
model=LogisticRegression()
```

```python
#training the LRM with the training dataset
model.fit(X_train,Y_train)
```

Accuracy score

```python
#accuracy of training data
X_training_data=model.predict(X_train)
training_data_accuracy=accuracy_score(X_training_data,Y_train)
```

```python
print("The training data accuracy is: ",training_data_accuracy)
```

```python
#accuracy of testing data
X_testing_data=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_testing_data,Y_test)
```

```python
print(f'The testing data accuracy is : {testing_data_accuracy}')
```
